# **There are two variants of this game. **

1.   In the first variant, you would have to play the 6 turns alloted to you.
2.   In the second variant, you have the option of exiting the game early.





```python
game_title = 'GUESS A 5-LETTER WORD'
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/python projects/words.txt', 'r') as file:
  word_bank = [line.strip() for line in file]
```

```python
print(f'Welcome to my not-so-simple {game_title} game')
print('Whether you win or loose, you only have 6 turns')
```

# **1ST VARIANT**

```python
your_turns = 6
turns_taken = 0

correct_guesses = []

while turns_taken < your_turns:
  prompt = input('Guess a word: ').lower()
  if len(prompt) != 5 or not prompt.isalpha():
    print('Please enter only words with 5 letters. No numbers, no characters')
    continue

  if prompt in word_bank:
    if prompt not in correct_guesses:
      correct_guesses.append(prompt)
      print('Great Job!, You guessed right!')
    else:
      print('You have guessed that word already, try another word')
  else:
    print('Try again, the word you guessed isnt in the word bank')

  turns_taken = turns_taken + 1
  turns_left = your_turns - turns_taken
  print('You have', turns_left, 'turns left')

  if turns_taken == your_turns:
    if len(correct_guesses) > 0:
      print('Game over! You guessed ', len(correct_guesses), 'words correctly:', {", ".join(correct_guesses)})
    else:
      print('You are out of turns, better luck next time')
```

## **2ND VARIANT**

```python
print('Welcome to my not-so-simple', game_title, 'game')
print('Whether you win or loose, you only have 6 turns')
print("If you wish to end the game, enter the word 'exit' when prompted")
```

```python
your_turns = 6
turns_taken = 0

correct_guesses = []

while turns_taken < your_turns:
  prompt = input("Guess a word OR type 'exit' to end the game: ").lower()

  if prompt == 'exit':
    print('Player has ended the game')
    break

  if len(prompt) != 5 or not prompt.isalpha():
    print('Please enter only words with 5 letters. No numbers, no characters')
    continue

  if prompt in word_bank:
    if prompt not in correct_guesses:
      correct_guesses.append(prompt)
      print('Great Job!, You guessed right!')
    else:
      print('You have guessed that word already, try another word')
  else:
    print('Try again, the word you guessed isnt in the word bank')

  turns_taken = turns_taken + 1
  turns_left = your_turns - turns_taken
  print('You have', turns_left, 'turns left')

  if turns_taken == your_turns:
    if len(correct_guesses) > 0:
      print(f'Game over! You guessed {len(correct_guesses)} words correctly: {", ".join(correct_guesses)}')
    else:
      print('You are out of turns, better luck next time')
```
